```python
from scipy.stats import ttest_1samp
#H0:Mean score of population = 984 (as mean score of sample 2 is 984)
#Ha:Mean score of population is not equal to 984

#confidence level=95%
alpha=0.05

expected_value = 984
t_statistic, p_value = ttest_1samp(df6['Score'], expected_value)

# Interpret the results
print(f"One-sample t-test results for 'Score' against {expected_value}:")
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("The 'Mean Score' is significantly different from 976 (reject the null hypothesis).")
else:
    print("The 'Mean Score' is not significantly different from 976 (fail to reject the null hypothesis).")
```

```python
import pandas as pd
from scipy.stats import f_oneway
# Research question: Is there a significant difference in the 'Score' among teams in different tournaments?

# Perform ANOVA test
tournaments = df3['Tournament'].unique()
anova_data = [df3[df3['Tournament'] == tournament]['Score'] for tournament in tournaments]
f_statistic, p_value = f_oneway(*anova_data)

# Define the significance level (alpha)
alpha = 0.05

# Interpret the results
print("ANOVA test results for 'Score' among different tournaments:")
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("There is a significant difference in 'Score' among teams in different tournaments (reject the null hypothesis).")
else:
    print("There is no significant difference in 'Score' among teams in different tournaments (fail to reject the null hypothesis).")
```
